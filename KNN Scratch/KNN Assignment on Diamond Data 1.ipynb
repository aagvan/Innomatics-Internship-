### Outliers Detection and Replace with Median

```python
df.info()
```

```python
plt.boxplot(df['carat'])
plt.title('weight of the diamond')
plt.xlabel('weight of the diamond')
plt.show()

def outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lb = q1 - (1.5*iqr)
    ub = q3 + (1.5*iqr)
    
    for x in df[col]:
        if x < lb or x > ub:
            df[col] = df[col].replace(x, np.median(df[col]))
        

outliers(df, "carat")        
plt.boxplot(df["carat"])
plt.show()
```

```python
plt.boxplot(df['depth'])
plt.title('depth of diamond')
plt.xlabel('depth in mm')
plt.show()

depth = outliers(df, "depth")        
plt.boxplot(df["depth"])
plt.show()
```

### Calculate Euclidean Distance between Two datapoints

```python
def euclidean_distance(row1, row2):
    
    distance = 0.0
    
    for i in range(len(row1)-1):    # here len(row)-1 means we are not taking y value
        distance += (np.sum((row1[i]- row2[i])**2))**0.5 # euclidean formula = ((Ai-bi)square)squareroot
        
    return distance    # here we taking squareroot
```

## Making a Model for KNN Regression from scratch

```python
class KNN_Regression:
    def __init__(self,k):
        self.k = k
        
    def fit_func(self,X,y):
        self.X = np.asarray(X)
        self.y = np.asarray(y)
        
    def predict_func(self,X):
        X = np.asarray(X)
        
        predict = []
        
        for i in X:
            distance = np.sqrt(np.sum((i - self.X)**2,axis=1))
            sort_value = distance.argsort()
            sort_value = sort_value[:self.k]
            predict.append(np.mean(self.y[sort_value]))
            
        return np.array(predict)
```

```python
Model = KNN_Regression(k=3)

Model.fit_func(X_train_transformed, y_train)

y_pred_KNN = Model.predict_func(X_test_transformed)
```

```python
from sklearn import metrics

knn_scratch = metrics.r2_score(y_test, y_pred_KNN)

knn_scratch
```
